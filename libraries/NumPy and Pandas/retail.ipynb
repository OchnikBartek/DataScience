```python
import numpy as np
import pandas as pd
```

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

```python
retail = retail_raw.copy()
retail.info()
```

```python
retail.describe().T
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
```

```python
retail.isnull().sum()
```

```python
retail = retail[retail.Quantity >= 0]
retail.info()
```

```python
retail.describe().T
```

```python
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
retail.info()
```

```python
retail[retail.CustomerID == '17850']
```

```python
retail[(retail.CustomerID == '17850') & (retail.UnitPrice > 5)]
```

```python
retail[retail.InvoiceNo.notnull()]
```

```python
retail.groupby('CustomerID').size()
```

```python
retail
```

```python
retail['Revenue'] = retail.Quantity * retail.UnitPrice
retail.head()
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': 'mean', 'CustomerID': 'size'}). rename(columns={'Revenue': 'RevenueAverage', 'CustomerID': 'CustomerIDCount'})
```

```python
retail.head(2)
```

```python
retail['InvoiceDateDay'] = retail.InvoiceDate.dt.day
retail.head(2)
```

```python
result = retail.groupby('InvoiceDateDay').aggregate({'Revenue' : 'sum'}).rename(columns={'Revenue': 'RevenueSum'})
retail.head(2)
```

```python
import seaborn as sns
sns.set_style()
result.plot(kind='bar')
```
