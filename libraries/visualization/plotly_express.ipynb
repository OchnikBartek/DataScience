```python
import plotly_express as px
```

```python
df = px.data.iris()
df
```

```python
px.scatter(data_frame=df, x = 'sepal_length', y = 'sepal_width',width=600, height=400, color='species',
           marginal_x= 'box', marginal_y= 'box',title='Iris', trendline='ols')
```

```python
px.scatter_matrix(data_frame=df,dimensions=['sepal_length','sepal_width','petal_length','petal_width'], color='species',title='Iris_matrix')
```

```python
px.parallel_coordinates(data_frame=df,color = 'species_id')
```

```python
def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df = df.reset_index()
df.head()
```

```python
px.area(data_frame=df, x='Date', y='Close')
```

```python
px.area(data_frame=df, x='Date', y='Close',log_y=True, title='Amazon')
```

```python
import pandas as pd
df = pd.read_csv("https://ml-repository-krakers.s3-eu-west-1.amazonaws.com/plotly-course/us-cities-top-1k.csv")
df.head()
```

```python
fig = px.scatter_map(df, lat='lat', lon='lon',
                        hover_name='City', hover_data=['State', 'Population'], zoom=3)


fig.update_layout(mapbox_style='carto-positron',
                    margin={'r': 10, 't': 10, 'l': 10, 'b': 10})
fig.show()
```
