```python
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('office.jpg')
plt.imshow(img)
_ = plt.axis('off')
```

```python
plt.figure(figsize=(12,10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'workplace.jpg', 'children.jpg']):
    plt.subplot(2, 2, idx+1)  # 2x2 siatka, idx+1 dla pozycji
    image = mpimg.imread(img)
    plt.imshow(image)
    plt.axis('off')  # Usunięcie osi

plt.tight_layout()  # Poprawia rozmieszczenie obrazów
plt.show()
```
